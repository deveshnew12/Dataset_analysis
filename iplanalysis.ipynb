```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ball=pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
match=pd.read_csv("IPL Matches 2008-2020.csv")
```

```python
match['year']=pd.DatetimeIndex(match['date']).year
```

```python
match.head()
```

```python
match_in_year=match.groupby(['year'])['id'].count().reset_index().rename(columns={'id':'total_match'})
match_in_year
```

```python
plt.bar(data=match_in_year,x='year',height='total_match')
```

```python
toss=match.groupby(['toss_winner'])['id'].count().reset_index().rename(columns={'id':'no_toss'})
```

```python
toss=pd.DataFrame(data=toss)
```

```python
toss
```

```python
plt.bar(data=toss,x='toss_winner',height='no_toss')
```

```python
chart=sns.barplot(data=toss,x='no_toss',y='toss_winner')
# chart.set_xticklabels(no_toss,data=toss)
```

```python
award=match.groupby(['player_of_match'])['id'].count().reset_index().rename(columns={'id':'no_award'})
```

```python
award
```

```python
award=award.sort_values('no_award',ascending=False)
```

```python
toptenplayer=award.head(10)
```

```python
sns.barplot(data=toptenplayer,x='no_award',y='player_of_match')
```

```python
match.head(2)
```

```python
toss_win=match.groupby(['toss_winner','toss_decision'])['id'].count().reset_index().rename(columns={'id':'no_decision'})
toss_win
```

```python
tot=toss_win.groupby(['toss_decision'])['no_decision'].sum().reset_index().rename(columns={'no_decision':'total'})
```

```python
plt.pie(data=tot,x='total',labels='toss_decision')
```
